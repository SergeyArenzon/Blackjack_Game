```python
import random


suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True

'''------------------Card------------------'''

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    
    def __str__(self):
        return self.rank+' of '+self.suit
    
 
'''------------------Deck------------------''' 

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        for card in self.deck:
            print(card)

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
    
 
'''------------------Hand------------------'''     

class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
            
    
    def adjust_for_ace(self):
        if self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1


'''------------------Chips------------------'''   

class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
    
    
'''------------------Take Bet func------------------'''    
def take_bet(chips):
    while True:
        while True:
            try:
                bet = int(input("Place a bet!"))
            except:
                print("Wrong bet")
            else:
                break
            
        if chips.total < bet:
            print("Not enough money!")
        else:
            chips.bet = bet
            break
            
            
'''------------------Hit func------------------'''  

def hit(deck,hand):
    card = deck.deal()
    hand.add_card(card) 
    hand.adjust_for_ace()
    
    
                    
'''------------------Hit or stand func------------------'''                 
        
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        action = input('Hit(1) or Stay(2)?')
        
        if action == '1':
            hit(deck,hand)
            break
            
        elif action == '2':
            print("Player stands, Dealers turn!")
            playing = False
            break
            
        else:
            print("Wrong input! choose 1 or 2")
            continue

'''------------------Show some func------------------''' 

def show_some(player,dealer):
    dealer_cards = ""
    for card in range(len(dealer.cards)):
        if card < len(dealer.cards)-1:
            dealer_cards += '['+str(dealer.cards[card])+']'
        else:
            dealer_cards += '[HIDDEN]'
            
    
    player_cards = ''
    
    for card in player.cards:
        player_cards += '['+str(card)+']'
    
    print(f'Dealer hand:{dealer_cards}\nPlayer hand:{player_cards}')
    
    
    
'''------------------Show All func------------------'''    

def show_all(player,dealer):
    dealer_cards = ""
    
    for card in dealer.cards:
        dealer_cards += '['+str(card)+']'
                
    player_cards = ''
    
    for card in player.cards:
        player_cards += '['+str(card)+']'
    
    print(f'Dealer hand:{dealer_cards}\n Player hand:{player_cards}')
   

'''------------------End game scenario func------------------''' 

def player_busts(player,dealer,chips):
    chips.lose_bet()
    print("BUST Player!")
    
    
def player_wins(player,dealer,chips):
    chips.win_bet()
    print('Player Wins!')

def dealer_busts(player,dealer,chips):
    chips.win_bet()
    print("Player Wins Dealer BUSTED")
    
def dealer_wins(player,dealer,chips):
    chips.lose_bet()
    print("Dealer Wins!")
    
def push(player,dealer):
    print("Player and Dealer Tie!")
    
    
    
'''------------------Game------------------'''   



while True:
   
    print("Welcome to Blackjack GAME!")
    

    
   
    deck = Deck()
    deck.shuffle()
    player_hand = Hand()
    dealer_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    
        
   
    chips = Chips()
    
    
    
    
    
    
    take_bet(chips)
    

    
   
    show_some(player_hand,dealer_hand)
    
    while playing:  
        
        
        hit_or_stand(deck,player_hand)
        
        

        show_some(player_hand,dealer_hand)
        
        
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,chips)
            break
            

    
    if player_hand.value <= 21:    
        while dealer_hand.value < 17:
            hit(deck,dealer_hand)
           
       
        show_all(player_hand,dealer_hand)
       
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,chips)
        
        elif  dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,chips)
        
        elif dealer_hand.value > player_hand.value:  
            dealer_wins(player_hand,dealer_hand,chips)
            
        else:
            push(player_hand,dealer_hand)
            
            
    
    print(f'Players Chips total is: {chips.total}')
    new_game = input("Do you want play again? Yes(1) no(2)")
    if new_game == '1':
        playing = True
        continue
    else:
        break
    

    break 

```
